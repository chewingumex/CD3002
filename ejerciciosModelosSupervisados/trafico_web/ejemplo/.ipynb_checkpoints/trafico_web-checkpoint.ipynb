```python
import pandas as pd
import datetime
import xgbRegression as xgbr
import xgbVisualisation as viz
import re
import xgboost
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# leer los datos

df = pd.read_csv('./../datos/daily-website-visitors.csv')
```

```python
# pre-procesar los datos

df['month'] = [datetime.datetime.strptime(x, "%m/%d/%Y").month for x in df.Date]
df['day_of_month'] = [datetime.datetime.strptime(x, "%m/%d/%Y").day for x in df.Date]

# quitar variables que no sirven para modelar
df = df.drop(['Row', 'Day', 'Date'],axis = 1)
```

```python
# arreglar nombres de las columnas

df.columns = [re.sub('\.','_',x.lower()) for x in df.columns]
```

```python
# quitar comas de las columnas numéricas

df = df.apply(lambda x: x.str.replace(',','').astype(int) if x.dtype == 'O' else x)
```

```python
# definir variables independientes y dependiente

yvar = df['unique_visits']
xvars = df[[x for x in df.columns if x != 'unique_visits']]
```

```python
# dividir en training y testing sets para las x y la y

xtrain, xtest, ytrain, ytest = train_test_split(xvars, yvar, train_size = 0.80, random_state = 2)
```

```python
# crear matrices de xgboost

trainXGB = xgbr.makeXGBMatrix(data=xtrain,
                             label=ytrain)

testXGB = xgbr.makeXGBMatrix(xtest,
                            ytest)
```

```python
# validacion cruzada y calibración de hiperparámetros
# ajuste de modelo final

finalXGB = xgbr.findXGBRegression(trainXGB,
                                 testXGB,
                                 xtrain,
                                 ytrain,
                                 xtest,
                                 ytest,
                                 iters=50)
```

```python
# visualizar el error durante el entrenamiento para ambas muestras

viz.visualise_metrics(finalXGB, mod = 'regression')
```

```python
# importancia de las variables en el modelo

viz.plotVariableImportance(finalXGB, viz_type=None)
```

```python
# contribucion de las variables a la predicción

viz.make_shap_visualisation(finalXGB, xtrain)
```
